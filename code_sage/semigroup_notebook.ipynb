```python
class semi_group:
    def __init__(self, c, g, m, G):
        self.c = c # conductor (max du complementaire +1)
        self.g = g # genus (cardinal complementaire)
        self.m = m # multiplicity (min S\{0})
        self.d = [0] * (G*3)
        
    def __str__(self):
        """ Pour pouvoir print """
        return "Conductor : {}, Genus : {}, Multiplicity : {}, Ds {}".format(self.c, self.g, self.m, self.d)

        
    def irreductible(self, x):
        """ Vrai si x est irreductible a l'addition de 2 elements du semigroup self """
        return self.d[x] == 1
            

def root(G):
    """ Racine de l'arbe Tg (les semi-groups avec genus inferieur ou egal a G"""
    R = semi_group(1, 0, 1, G) #(0, 0, 1, G)
    for x in range(3 * G):
        R.d[x] = 1 + (x//2)
    return [R]

def son(S, x, G):
    """ fonction successeur """
    if x > S.m:
        m = S.m
    else:
        m = S.m + 1
    Sx = semi_group(x + 1, S.g + 1, m, G)
    Sx.d = S.d

    for y in range(x, (3*G)+1): #+1 car 3G inclus
        if S.d[y - x] > 0:
                Sx.d[y] = S.d[y] - 1
                print(Sx.d[y - x])
                print(Sx.d[y])
    return Sx

G_test = 4
seeds = root(G_test)
#succ = lambda m: []
def succ(m):
    r = []
    for x in range(3 * G_test):
        # if x est irreductible a l'addition de 2 elements de m :
        if m.irreductible(x):
            #print(x)
            r.append(son(m, x, G_test))
    return r
        
Tg = RecursivelyEnumeratedSet(seeds, succ, structure='forest')
```
